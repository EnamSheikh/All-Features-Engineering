# One Hot Encoding

# Ordinal Encoding

```python
df = pd.read_csv('agora.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df.Area.unique()
```

```python
city_list = ['Dhaka', 'Ctg', 'Rangpur']
```

```python
ordinal = OrdinalEncoder(categories=[city_list])
```

```python
encoded_variables = ordinal.fit_transform(df[['Area']])
```

```python
new_area = pd.DataFrame(encoded_variables, columns =['Area'])
```

```python
new_area.head()
```

```python
new_df = df.drop('Area', axis = 1)
```

```python
pd.concat([new_df,new_area], axis = 1).head()
```
